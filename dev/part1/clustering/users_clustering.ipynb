```python
import pandas as pd
import numpy as np
from scipy import sparse
from sklearn.preprocessing import normalize
from tqdm.notebook import tqdm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import math
import seaborn as sns
import pickle
from sklearn.cluster import DBSCAN, KMeans
from collections import Counter
```

```python
train_joke_df = pd.read_csv(r'data\recsys-in-practice\train_joke_df.csv')
```

```python
npusers_arr = np.zeros((24983, 100), dtype=np.half)

for row in tqdm(train_joke_df.values):
    npusers_arr[int(row[0]) - 1, int(row[1]) - 1] = row[2]
```

```python
%%time

kmeans = KMeans(300, random_state=42)
labels = kmeans.fit_predict(npusers_arr)


count = Counter(labels)
len(count), sorted(count.items(), key=lambda item: item[1], reverse=True)
```

```python
%%time

kmeans = KMeans(1000, random_state=42)
labels = kmeans.fit_predict(npusers_arr)

count = Counter(labels)
len(count), sorted(count.items(), key=lambda item: item[1], reverse=True)[:10]
```

```python
%%time

kmeans = KMeans(5000, random_state=42)
labels = kmeans.fit_predict(npusers_arr)

count = Counter(labels)
len(count), sorted(count.items(), key=lambda item: item[1], reverse=True)[:20]
```

```python
%%time

kmeans = KMeans(10000, random_state=42)
labels2 = kmeans.fit_predict(npusers_arr)

count = Counter(labels2)
len(count), sorted(count.items(), key=lambda item: item[1], reverse=True)[:20]
```

```python
%%time

kmeans = KMeans(20000, random_state=42)
labels3 = kmeans.fit_predict(npusers_arr)

count = Counter(labels3)
len(count), sorted(count.items(), key=lambda item: item[1], reverse=True)[:20]
```

```python
cluster_df = pd.DataFrame({'cluster_id':labels3, 'UID':range(1, npusers_arr.shape[0] + 1)})
cluster_df
```

```python
train_joke_df.merge(cluster_df)
```

```python
cluster_df.to_csv('cluster_df.csv', index=False)
```
